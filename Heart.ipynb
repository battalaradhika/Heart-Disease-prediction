```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
X = 250  # Price per prediction in Rs.
r = 0.032  # Growth rate (3.2%)
months = np.array([1, 2, 3, 4, 5, 6])  # Time in months
sales = 300  # Number of units sold each month

# Calculate profit over time
profits = sales * X * (1 + r)**months

# Plot the results
plt.plot(months, profits, marker='o', color='blue', label='Profit Over Time')
plt.xlabel('Month')
plt.ylabel('Profit (Rs.)')
plt.title('Heart Disease Prediction System Profit Forecast')
plt.legend()
plt.grid(True)
plt.show()
```
